def connecter_sql(nom_du_archive):
    bd = sqlite3.connect(nom_du_archive)
    cursor = bd.cursor()
    return cursor

# fonction pour effectuer une requête
def effectuer_requete(requete): #requete é q funcao sql SELECT etc
    l = []
    for elem in cursor.execute(requete):
        l.append(elem)
    return l

# fonction pour récupérer le résultat 
def recuperer_resultat(requete):
    cursor.execute(requete)
    resultat = cursor.fetchall()
    return resultat

# fonction pour insérer de données via pandas 
def insertion_replace(name, conn, dataframe):
    insertion = dataframe.to_sql(name, con=conn, if_exists='replace', index = False)
    return insertion

```python
# so para um valor por coluna
def gen_insert_query(table_name:str, a_dict:dict) -> str:
    return f"INSERT INTO {table_name} {tuple(a_dict.keys())} VALUES {tuple(a_dict.values())}"
```

```python
def gen_insert_queries(table_name:str, a_dict:dict) -> str:
    return f"INSERT INTO {table_name} {tuple(a_dict.keys())} VALUES {','.join(str(i) for i in tuple(zip(*a_dict.values())))}"
```

```python
def gen_insert_queries_df(table_name:str, df:pd.DataFrame) -> str:
    dico = df.to_dict(orient='list')
    return f"INSERT INTO {table_name} {tuple(dico.keys())} VALUES {','.join(str(i) for i in tuple(zip(*dico.values())))}"
```
