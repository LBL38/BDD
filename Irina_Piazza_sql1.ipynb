```python
import pandas as pd
import matplotlib.pyplot as plt
conn_pd = sqlite3.connect('billboard/billboard-200.db') 
          
sql_query = pd.read_sql_query ('''
                               SELECT
                               *
                               FROM acoustic_features
                               ''', conn_pd)

df = pd.DataFrame(sql_query)
print (df)
```

```python
df
```

```python
df.describe() #mean
```

```python
df['date'] =  pd.to_datetime(df['date'], format='%Y-%m-%d')
df['year'] = df['date'].dt.year
df = df.set_index('year')
df
```

```python
df.groupby('year').mean().plot()
```

```python
df.groupby('year').describe()
```

```python
test =df.groupby('year')
test["loudness"].mean().max()
```

```python
means=df.groupby('year').mean().drop('key', axis=1)
with plt.style.context('fivethirtyeight'):
    fig, axes = plt.subplots(3, 4, figsize=(25, 15), sharex=True)
    for i, col in enumerate(means):
        axes.flatten()[i].plot(means[col])
        axes.flatten()[i].set_title(col)
        
```
